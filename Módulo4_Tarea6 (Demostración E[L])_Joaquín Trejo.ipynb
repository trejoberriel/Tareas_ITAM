# Tarea 6

Demostrar que :

$$ E[\;L\;(t,y(\bar{x}))\;] = \displaystyle\int (\;{[\;y(\bar{x})\;-\;E(t|\bar{x})\;]}^{2}+V_t(t|\bar{x})\;)\;P(\bar{x})\;dx$$

Sabiendo que  $$ 1) L(\;t,y(\bar{x})\;) = {(\;y(\bar{x})-t\;)}^{2}\;\;,\;\;2)\;\;y(\bar{x}) = E[t|\bar{x}]$$

# Demostraciòn:

Por definiciòn:
    
$$ E[L(t,y(\bar{x}))\;] = \displaystyle\int\displaystyle\int  L(t,y(\bar{x}))\;P(t,\bar{x})\;dt\;d\bar{x}$$

Ahora bien, considerando 1):

$$ E[L(t,y(\bar{x}))\;] = \displaystyle\int\displaystyle\int  {(\;y(\bar{x})-t\;)}^{2}\;P(t,\bar{x})\;dt\;d\bar{x}$$

Podemos sumar y restar el término 2) al término cuadràtico:

$$ E[L(t,y(\bar{x}))\;] = \displaystyle\int_{\bar{x}}\displaystyle\int_{t}  {(\;y(\bar{x})\;-\;E[t|\bar{x}]\;+\;E[t|\bar{x}]\;-\;t\;)}^{2}\;P(t,\bar{x})\;dt\;d\bar{x}$$

Desarollamosel término cuadrático:

$$E[L(t,y(\bar{x}))\;] = \displaystyle\int_{\bar{x}}\displaystyle\int_{t}[\;{(y(\bar{x})\;-\;E[t|\bar{x}] \;)}^{2}\;+\;2(\;y(\bar{x})\;-\;E[t|\bar{x}] \;)(\;E[t|\bar{x}]\;-\;t)\;)\;+\;{(\;E[t|\bar{x}] \;-t\;)}^{2}\;]\;P(t,\bar{x})\;dt\;d\bar{x}$$


Procedemos a desarrollar cada 1 de las 3 integrales. Por definición sabemos que:

$$\;P(t|\bar{x})\;=\;\frac{P(t,\bar{x})}{P(\bar{x}}$$


**Primera integral**

$$ \displaystyle\int_{\bar{x}}\displaystyle\int_{t}\;{(y(\bar{x})\;-\;E[t|\bar{x}] \;)}^{2}\;P(t,\bar{x})\;dt\;d\bar{x}$$

$$\; = \;\displaystyle\int_{\bar{x}}\displaystyle\int_{t}\;{(y(\bar{x})\;-\;E[t|\bar{x}] \;)}^{2}\;P(t|\bar{x})P(\bar{x})\;dt\;d\bar{x}$$

Podemos sacar de la integral sobre t, aquellos tèrminos que sean constantes:

$$\; = \;\displaystyle\int_{\bar{x}}\;P(\bar{x})\;\displaystyle\int_{t}\;{(y(\bar{x})\;-\;E[t|\bar{x}] \;)}^{2}\;P(t|\bar{x})\;dt\;d\bar{x}$$

Ahora bien, el siguiente tèrmino:

$$\displaystyle\int_{t}\;{(y(\bar{x})\;-\;E[t|\bar{x}] \;)}^{2}\;P(t|\bar{x})\;dt\;$$

por definición es:

$$E[\;{(\;y(\bar{x})\;-\;E[t|\bar{x}] \;)}^{2}|\;\bar{x}\;],\;sobre\;t$$

y el cual se puede reducir a:

$$({y(\bar{x})\;-\;E[t|\bar{x}] \;)}^{2}\;$$

Por lo tanto nuestra primer integral toma la siguiente forma:

$$\; = \;\displaystyle\int_{\bar{x}}\;(\;{y(\bar{x})\;-\;E[t|\bar{x}] \;)}^{2}\;P(\bar{x})\;d\bar{x}$$

**Segunda integral**

$$\displaystyle\int_{\bar{x}}\displaystyle\int_{t}\;2(\;y(\bar{x})\;-\;E[t|\bar{x}] \;)(\;E[t|\bar{x}]\;-\;t\;)\;P(t,\bar{x})\;dt\;d\bar{x}$$

$$\;=\;\displaystyle\int_{\bar{x}}\displaystyle\int_{t}\;2(\;y(\bar{x})\;-\;E[t|\bar{x}] \;)(\;E[t|\bar{x}]\;-\;t\;)\;\;P(t|\bar{x})P(\bar{x})\;dt\;d\bar{x}$$

Podemos sacar de la integral sobre t, aquellos tèrminos que sean constantes:


$$\;=\;\displaystyle\int_{\bar{x}}\;2P(\bar{x})\displaystyle\int_{t}\;(\;y(\bar{x})\;-\;E[t|\bar{x}] \;)(\;E[t|\bar{x}]\;-\;t\;)\;\;P(t|\bar{x})\;dt\;d\bar{x}$$

Ahora bien, el siguiente tèrmino:

$$\displaystyle\int_{t}\;(\;y(\bar{x})\;-\;E[t|\bar{x}] \;)(\;E[t|\bar{x}]\;-\;t\;)\;\;P(t|\bar{x})\;dt$$


por definición es:

$$E[\;(\;y(\bar{x})\;-\;E[t|\bar{x}] \;)(\;E[t|\bar{x}]\;-\;t\;)\;|\;\bar{x}\;],\;sobre\;t$$

y el cual se puede reducir a 0 pues:

$$E[\;(E[t|\bar{x}\;]\;-\;t)\;|\;\bar{x}]\;=\;0\;,\;\;sobre\;t$$

por lo tanto, toda nuestra segunda integral se reduce a **0**.


**Tercera intergal**

$$\displaystyle\int_{\bar{x}}\displaystyle\int_{t}\;{(\;E[t|\bar{x}] \;-t\;)}^{2}\;\;P(t,\bar{x})\;dt\;d\bar{x}$$

$$=\;\displaystyle\int_{\bar{x}}\displaystyle\int_{t}\;{(\;E[t|\bar{x}] \;-t\;)}^{2}\;\;P(t|\bar{x})P(\bar{x})\;dt\;d\bar{x}$$

Podemos sacar de la integral sobre t, aquellos tèrminos que sean constantes:

$$=\;\displaystyle\int_{\bar{x}}P(\bar{x})\displaystyle\int_{t}\;{(\;E[t|\bar{x}] \;-t\;)}^{2}\;\;P(t|\bar{x})\;dt\;d\bar{x}$$

Ahora bien, el siguiente tèrmino:

$$\displaystyle\int_{t}\;{(\;E[t|\bar{x}] \;-t\;)}^{2}\;\;P(t|\bar{x})\;dt$$

se puede visualizar como:

$$\displaystyle\int_{t}\;{(\;t \;-E[t|\bar{x}]\;)}^{2}\;\;P(t|\bar{x})\;dt$$

ya que:

$${(a+b)}^{2} = {(a-b)}^{2}$$

Ahora bien, por definición:


$$\displaystyle\int_{t}\;{(\;t \;-E[t|\bar{x}]\;)}^{2}\;\;P(t|\bar{x})\;dt$$

es la varianza condicional de  t respecto a x barra, la cual se simboliza de la siguiente manera:

$$V_t(t|\bar{x})\;$$

por lo que nuestra tercera integral queda de la siguiente forma:

$$=\;\displaystyle\int_{\bar{x}}\;V_t(t|\bar{x})\;P(\bar{x})\;d\bar{x}$$


Sumando la **integral 1** y la **integral 2**:

$$\; = \;\displaystyle\int_{\bar{x}}\;(\;{y(\bar{x})\;-\;E[t|\bar{x}] \;)}^{2}\;P(\bar{x})\;d\bar{x}\;+\;\displaystyle\int_{\bar{x}}\;V_t(t|\bar{x})\;P(\bar{x})\;d\bar{x}$$

queda demostrado el ejercicio:

$$ E[\;L\;(t,y(\bar{x}))\;] = \displaystyle\int (\;{[\;y(\bar{x})\;-\;E(t|\bar{x})\;]}^{2}+V_t(t|\bar{x})\;)\;P(\bar{x})\;dx$$
